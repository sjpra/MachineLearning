## Lets finally get to predicting?. 

*** I Decide to Analyse using KNN classification method ***

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df_knn = df_new
df_knn = df_knn.drop(['first_funding_at','last_funding_at'], axis=1)
df_knn.head()
```

*** Some Transformation so that we can do the analysis ***

```python
df_knn['success'] = df_knn['status']
df_knn = df_knn.drop(['status'], axis=1)
df_knn.head(10)
```

```python
df_new.isnull().sum()
```

*** Lets just check all the types of the columns once as we can not use KNN on non int/ float feilds ***

```python
df_knn.info()
```

```python
scaler.fit(df_knn.drop('success',axis=1))
scaled_features = scaler.transform(df_knn.drop('success',axis=1))
df_feat = pd.DataFrame(scaled_features,columns=df_knn.columns[:-1])
df_feat.head()
```

## Lets bring the test/ train friend up

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features,df_knn['success'],
                                                    test_size=0.30)
```

## Lets first try with K = 1 and see what happens

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

*** Lets Evaluate what is happening ***

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

## Okay that's around 60 Percent. Lets see if we can increase this percentage by choosing a better K value

```python
error_rate = []

# Will take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

## I choose 16!

```python
knn = KNeighborsClassifier(n_neighbors=16)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=16')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
